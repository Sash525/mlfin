План такой:
Загружаем по отдельности датасеты и присоединяем их к общему.
Поля будут одинаковые, будем менять только преффиксы
Никаких вычислений с полями делать не будем, только данные

```python
import numpy as np
import pandas as pd
```

```python

def get_data(datafile, iday, preff):
        ''' Функция считывает данные из csv файла МТ5
        '''
        
        if iday == True:
            raw = pd.read_csv(datafile,
            encoding='utf-8',
            index_col=False,
            skiprows=1,
            header= None,
            names= [preff + '_date',
            preff + '_time',
            preff + '_open',
            preff + '_high',
            preff + '_low', 
            preff + '_close', 
            preff + '_tv', 
            preff + '_vol', 
            preff + '_spread'],
            sep='\t')
            
            raw['datetime'] = raw[preff + '_date'] + ' ' + raw[preff + '_time']
            raw.reset_index(drop=True, inplace=True)
            raw = raw.set_index('datetime')
            raw.drop([preff + '_date', preff + '_time', preff + '_vol'], axis=1, inplace=True)
            
            return raw
        
        else:
            raw = pd.read_csv(datafile,
            encoding='utf-8',
            index_col=0,
            skiprows=1,
            header= None,
            parse_dates=True,
            names= ['date',
            preff + '_open',
            preff + '_high',
            preff + '_low', 
            preff + '_close', 
            preff + '_tv', 
            preff + '_vol', 
            preff + '_spread'],
            sep='\t')
            
            return raw
```

```python
# Составим список с тикерами
preff_list = ['AUDJPY', 'AUDNZD', 'AUDUSD', 'EURAUD', 'EURCHF', 'EURGBP', 'EURJPY',
'EURUSD', 'GBPCHF', 'GBPJPY', 'GBPUSD', 'NZDUSD', 'USDCAD', 'USDCHF', 'USDJPY']
```

```python
data = pd.DataFrame()
# datafile = '/home/sash/Python/projects/mlfin/data/currency_1H/AUDNZD_H1_202011012200_202111010000.csv'
iday = True
for preff in preff_list:
    datafile = '/home/sash/Python/projects/mlfin/data/currency_1H/' + preff + '_H1_202011012200_202111010000.csv'
    data = pd.concat([data, get_data(datafile, iday, preff)], axis=1)
```
